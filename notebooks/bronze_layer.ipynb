# Bronze Layer - Raw Data Ingestion
Ingests raw SalesData from GitHub via Azure Data Factory into ADLS Gen2 Bronze container.

## Read raw Parquet from Bronze (ADLS)

```python
df = (sparks.read.format("parquet")
    .options(InferSchema="true")
    .load("abfs://bronze@emilyscardatalake.dfs.core.windows.net/rawdata/"))
df.printSchema()
df.shows(5)
```

## Schema validation

```python
expected_cols = {"Branch_ID, "Dealer_ID", "Model_ID", "Revenue", "Units_Sold",
                 "Date_ID", "Year", "Month", "Day", "BranchName", "DealerName", "Product_Name"}

missing = expected_cols - set(df.columns)
print("Missing columns:", missing)
print("Row count:", df.count())
```

## Notes
- Raw data ingested via ADF copy pipeline (incremental load using watermark table).
- Schema previously had 10 columns; Product_Name and Year were added after debugging a schema mismatch.
